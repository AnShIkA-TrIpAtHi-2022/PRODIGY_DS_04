```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
# Load the CSV file
file_path = "dataset_task_04.csv"
data = pd.read_csv(file_path)
```

```python
# Renaming columns for clarity
data.columns = ['ID', 'Brand', 'Sentiment', 'Post']

# Dropping the 'ID' column as it's not necessary for sentiment analysis
data = data.drop(columns=['ID'])

# Checking the distribution of sentiment labels
sentiment_distribution = data['Sentiment'].value_counts()
```

```python
# Plotting the distribution of sentiment labels
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='Sentiment', order=sentiment_distribution.index, palette='viridis')
plt.title('Distribution of Sentiment Labels')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
# Plotting sentiment distribution by brand
plt.figure(figsize=(14, 8))
sns.countplot(data=data, x='Brand', hue='Sentiment', palette='viridis', dodge=True)
plt.title('Sentiment Distribution by Brand')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Sentiment')
plt.show()
```

```python
# Function to generate word cloud
def generate_word_cloud(text, title):
    wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='viridis').generate(text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(f'Word Cloud for {title} Sentiment')
    plt.axis('off')
    plt.show()
```

```python
# Generate text for each sentiment category
neutral_text = ' '.join(data[data['Sentiment'] == 'Neutral']['Post'])
positive_text = ' '.join(data[data['Sentiment'] == 'Positive']['Post'])
negative_text = ' '.join(data[data['Sentiment'] == 'Negative']['Post'])
irrelevant_text = ' '.join(data[data['Sentiment'] == 'Irrelevant']['Post'])
```

```python
# Generate word clouds
generate_word_cloud(neutral_text, 'Neutral')
generate_word_cloud(positive_text, 'Positive')
generate_word_cloud(negative_text, 'Negative')
generate_word_cloud(irrelevant_text, 'Irrelevant')
```
